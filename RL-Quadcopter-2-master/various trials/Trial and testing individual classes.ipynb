```python
import sys
```

```python
initial_pos=np.array([0.0, 0.0, 10.0,0.0,0.0,0.0])
target_pos=np.array([0.0,0.0,50.0])
runtime=5
task=Task(init_pose=initial_pos, runtime=runtime, target_pos=target_pos)
```

```python
sample_size=128
buffer_size=200000
tau=0.7
learning_rate=0.001
agent=Agent(buffer_size=buffer_size, env=task, tau=tau, sample_size=sample_size, learning_rate=0.01)
```

```python
agent.local_actor.model.summary()
```

```python
agent.local_critic.model.summary()
```

```python
num_episodes = 2000
scores=[]
test_scores=[]
avg_pos=[]
test_avg_pos=[]
```

```python
for episode in range(1,num_episodes+1):
    state=agent.reset()
    done = False
    while not done:
        action=agent.action(state=state)
        next_state, reward, done = agent.step(state=state,action=action)
        if done:
            scores.append(agent.score)
            avg_pos.append(agent.average_pos)
            print("\rEpisode: {}, current_score:{:.2f}, best_score:{:.2f}, noise_scale:{:.4f}, agent_learning: {}".format(episode, agent.score, agent.best_score, agent.noise_scale, agent.learning_mode), end=" ")
        state=next_state
    sys.stdout.flush() 
```

```python
import matplotlib.pyplot as plt
```

```python
def n_episode_avg(n=10):
    n_episode_avg=[]
    for iter in range(len(scores)-n):
        avg=np.array(scores[iter:iter+n]).mean()
        n_episode_avg.append(avg)
    plt.figure()
    plt.plot(n_episode_avg)
    plt.ylabel('avg_score_over_{}_episodes'.format(n))
    plt.xlabel('episodes')
    plt.show()    
```
